```python
from torch.utils.data import Dataset, DataLoader
import os
import pydicom
import torch
import torchvision
import torchvision.transforms as transforms
from PIL import Image

import pylibjpeg

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2

root_dir = 'train'
image_labels = pd.read_csv('C:/designcredit/designcredit/vindr/image_labels_train.csv')

# empty train_png folder and thne create new one
if os.path.exists('train_png'):
    os.system('rm -rf train_png')
os.system('mkdir train_png')

for idx in range(image_labels.shape[0]):
    if idx % 100 == 0:
        print(idx)
    image_file = pydicom.dcmread(os.path.join(root_dir, f'{image_labels.iloc[idx, 0]}.dicom'))
    # plt.imshow(image_file.pixel_array, cmap=plt.cm.bone)
    image = image_file.pixel_array.astype(float)
    image = image / 4096                           # scaling from 0-1
    image = (image * 255).astype(np.uint8)    # converting to uint8 and scaling to 0-255
    # image = Image.fromarray(image)
    # plt.imshow(image, cmap=plt.cm.bone)
    # plt.show()
    image = cv2.resize(image, dsize=(224, 224), interpolation=cv2.INTER_CUBIC)
    cv2.imwrite(f'{root_dir}_png/{image_labels.iloc[idx, 0]}.png', image)
    
    # #print size of both the dicom and png file in kilobytes
    # print(f'{image_labels.iloc[idx, 0]}.dicom: {os.path.getsize(os.path.join(root_dir, f"{image_labels.iloc[idx, 0]}.dicom"))/1000} KB')
    # print(f'{image_labels.iloc[idx, 0]}.png: {os.path.getsize(os.path.join(f"{root_dir}_png", f"{image_labels.iloc[idx, 0]}.png"))/1000} KB')
    


    # label = image_labels.iloc[idx, 2:].astype(float)
    # label = torch.from_numpy(np.array(label))
    # # print(image.shape, label.shape)
    # # print(type(image), type(label))
    # return image, label
```
